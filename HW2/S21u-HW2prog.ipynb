### Programming problem 1
Recall that in problem 1, give a recursive algorithm to compute the product of two positive integers, $n$ and $m$, using only addition and subtraction.
This programming problem is to consider the recursive approach in above problem and asks you to complete the following tasks.
1. Implement the approach as a function named as `product_rec` using Python.
2. Please have an iterative version for the approach and write a function for this version with function name `product_ite`.
3. Compare these two function with the same input in terms of running time and write what you have observed.

Note that we will use an in-built python library **`timeit`** and the module function **`timeit.timeit()`** for measuring the running time.

```python
# Please write your code for product_rec() here
def product_rec(a, b):
    if a < b:
        return product_rec(b, a) 
    elif b == 0:
        return 0
    else:
        return a + product_rec(a, b-1)

# print(product_rec(100, 100))
    
```

```python
# Please write your code for product_ite()
def product_ite(a, b):
    product = 0
    if a < b :
        return product_ite(b, a)
    for i in range(b):
        product += a
    return product

# print(product_ite(100, 100))
```

```python
# This piece of code is for the ways to measure the running time.
# Please check the document of timeit library for more detail.
# We here briefly give two methods for your reference.
#---------------------------------------------------------timeit
import timeit as ti

#--------------------------- function for illustration
def product_rec(a, b):
    if a < b:
        return product_rec(b, a) 
    elif b == 0:
        return 0
    else:
        return a + product_rec(a, b-1)

def product_ite(a, b):
    product = 0
    if a < b :
        return product_ite(b, a)
    for i in range(b):
        product += a
    return product

#---------------------------input
x = int(input("Enter X:"))
y = int(input("Enter Y:"))
r = int(input("Enter runs:"))

#---------------------------------------------------------Method 1
print("Running time comparison (Timer Method 1):")
tstart = ti.default_timer() #timer start 計時器開啟
for i in range(r):
    product_rec(x, y)
tend = ti.default_timer() #timer end計時器關閉
t_rec=tend - tstart#開始與結束的差值

tstart = ti.default_timer() #timer start 計時器開啟
for i in range(r):
    product_ite(x, y)
tend = ti.default_timer() #timer end計時器關閉
t_ite=tend - tstart#開始與結束的差值
print("Recursion result", product_rec(x, y),"Iteration result", product_ite(x, y))
print("Recursive approach:", t_rec)
print("iterative approach:", t_ite)

#---------------------------------------------------------Method 2
print("Running time comparison (Timer Method 2):")
t_rec = ti.timeit("product_rec(259, 767)", setup="from __main__ import product_rec", number = r)#timeit(函數名稱,執行次數)
t_ite = ti.timeit("product_ite(259, 767)", setup="from __main__ import product_ite", number = r)

print("Recursion result", product_rec(x, y),"Iteration result", product_ite(x, y))
print("Recursive approach:", t_rec)
print("Iterative approach:", t_ite)
```
